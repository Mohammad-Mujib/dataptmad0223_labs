#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
url
````

````python
# your code here
response = requests.get(url)
response = response.content
type(response)
````

````python
soup = BeautifulSoup(response, 'html.parser')
soup
````

````python
repo_tags = soup.find_all('h1', {'class': 'h3 lh-condensed'})
repo_tags
````

````python
repos = []
for tag in repo_tags:
    repo = tag.text.strip().replace('\n', '').split('/')
    repos.append({'owner': repo[0], 'name': repo[1]})
````

````python
print(repos)
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# your code here
res4 = requests.get(url4)
res4 = res4.content
type(res4)
````

````python
soup4 = bs4.BeautifulSoup(res4, 'html.parser')
soup4
````

````python
links = soup4.find_all('a')
links
````

````python
link_list = []
for link in links:
    href = link.get('href')
    if href and href.startswith('/wiki/'):
        link_list.append('https://en.wikipedia.org' + href)
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.emsc-csem.org/Earthquake/'
url7
````

````python
# your code here
res7 = requests.get(url7)
res7 = res7.content
type(res7)
````

````python
soup7 = bs4.BeautifulSoup(res7, 'html.parser')
soup7
````

````python
table = soup7.find_all('table', {'id': 'tabearthquakes'})
table
````

````python
data = []
for row in table.tbody.find_all('tr', {'class': 'ligne1, ligne2'}):
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    date_time = cols[3].split()
    data.append({
        'date': date_time[0],
        'time': date_time[1],
        'latitude': cols[4],
        'longitude': cols[5],
        'region': cols[11]
    })
````
