## Challenge 1

#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1

# create two subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# plot (x, y) on the first subplot
ax1.plot(x, y, linewidth=2, linestyle='--')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('y = 2x')

# plot (x, z) on the second subplot
ax2.plot(x, z, linewidth=2, linestyle='-.')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_title('z = x^2')

# adjust the spacing between subplots
fig.subplots_adjust(hspace=0.5)

# display the plot
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
# create a single plot with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 8))

# plot (x, y) on the first subplot
ax1.plot(x, y, linewidth=2, linestyle='--')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('y = 2x')

# plot (x, z) on the second subplot
ax2.plot(x, z, linewidth=2, linestyle='-.')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_title('z = x^2')

# display the plot
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
# Create figure and axis objects
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8,10))

# Plot data on the axis objects
axs[0].plot(x, y, linewidth=2, linestyle='--', color='blue', label='y = 2x')
axs[1].plot(x, z, linewidth=2, linestyle='-', color='red', label='z = x^2')

# Set axis labels and title
axs[0].set_xlabel('x', fontsize=14, fontweight='bold')
axs[0].set_ylabel('y', fontsize=14, fontweight='bold')
axs[0].set_title('Plot of y = 2x', fontsize=16, fontweight='bold')
axs[1].set_xlabel('x', fontsize=14, fontweight='bold')
axs[1].set_ylabel('z', fontsize=14, fontweight='bold')
axs[1].set_title('Plot of z = x^2', fontsize=16, fontweight='bold')

# Add legend
axs[0].legend()
axs[1].legend()

# Show the plot
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

fitbit.groupby('Months_encoded')['Steps'].median().plot(ax=ax1)
ax1.set_xticks(range(1,13))
ax1.set_xlabel('Month')
ax1.set_ylabel('Median Steps')
ax1.set_title('Median Steps by Month')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

fitbit.groupby('Work_or_Weekend')['Steps'].median().plot(ax=ax2, kind='bar')
ax2.set_xticklabels(['Weekend', 'Workday'])
ax2.set_xlabel('')
ax2.set_ylabel('Median Steps')
ax2.set_title('Median Steps by Workday/Weekend')

# Save figures to figures folder
#fig.savefig('figures/fitbit_visualizations.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for i, feature in enumerate(features):
    ax = axes[i]
    ax.scatter(fitbit[feature], fitbit['Steps'], alpha=0.5)
    ax.set_xlabel(feature)
    ax.set_ylabel('Steps')
    ax.set_title(f'{feature} vs Steps')

plt.tight_layout()
plt.show()
```
